```python
import os
import json
import numpy as np
import pandas as pd
from natsort import index_natsorted, order_by_index
```

```python
FILE_PATH = './data/retweets/Eng/'
files = os.listdir(FILE_PATH)
retweet = pd.read_csv(FILE_PATH+files[0], error_bad_lines=False,)
for i in range(1,len(files)):
    temp = pd.read_csv(FILE_PATH+files[i], error_bad_lines=False)
    retweet = pd.concat([retweet, temp], ignore_index=True)
```

```python
retweet.shape
```

```python
df = retweet.reindex(index=order_by_index(retweet.index, index_natsorted(retweet['user_id'])))
df.head()
```

```python
df = retweet.drop_duplicates(subset='id_str', keep='last')
df = df.reindex(index=order_by_index(df.index, index_natsorted(df['user_id'])))
df.head()
```

```python
df.to_csv('./output/filtered/retweet_text.csv',sep=',',index=False)
```

```python
TWEET_PATH = './output/filtered/'
COMBINED = '/combined_vaccine.csv'
MASK = '/masking.csv'
filtered_files = os.listdir(TWEET_PATH)
vaccine_file =  pd.read_csv(TWEET_PATH + filtered_files[0] + COMBINED)
mask_file =  pd.read_csv(TWEET_PATH + filtered_files[0] + MASK)
for i in range(1,len(filtered_files)):
    temp1 = pd.read_csv(TWEET_PATH+filtered_files[i]+COMBINED, error_bad_lines=False)
    temp2 = pd.read_csv(TWEET_PATH+filtered_files[i]+MASK, error_bad_lines=False)
    vaccine_file = pd.concat([vaccine_file, temp1], ignore_index=True)
    mask_file = pd.concat([mask_file, temp2], ignore_index=True)
```
