```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mask = pd.read_csv('BERT_mask_result.csv')
vaccine = pd.read_csv('BERT_vaccine_result.csv')
```

```python
data = mask
cluster = list(data['cluster'])
x = list(data['x'])
y = list(data['y'])
```

```python
df = pd.DataFrame(dict(x=x, y=y, cluster=cluster))
groups = df.groupby('cluster')

fig, ax = plt.subplots()
ax.margins(0.05)
for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=0.5, label=name)

ax.legend()
plt.title('BERT Cluster Result - Mask')
plt.show()
```

```python
for name, group in groups:
    print(str(name) + ": " + str(len(group)))
```
