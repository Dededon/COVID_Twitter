```python
import re
import pandas as pd
```

```python
df = pd.read_csv('.\data\cluster_result_features.csv')
location = list(df['location'])
```

```python
cities = pd.read_csv(".\city_data\city_states.csv", encoding='latin1')
counties = pd.read_csv(".\city_data\counties_states.csv", encoding='latin1')
states = pd.read_csv(".\city_data\states_codes.csv", encoding='latin1')
```

```python
cities = cities.fillna('0')

cities['state'] = cities['state'].str.lower()
states['state'] = states['state'].str.lower()
counties['state'] = counties['state'].str.lower()


cities['city'] = cities['city'].str.lower()
states['code'] = states['code'].str.lower()
counties['county'] = counties['county'].str.lower()
```

```python
cities = pd.merge(cities,states, on='state')
counties = pd.merge(counties,states, on='state')

cities_to_state_code_mapping =  cities.set_index('city').T.to_dict()
state_to_code_mapping = states.set_index('state').T.to_dict()
counties_to_state_code_mapping = counties.set_index('county').T.to_dict()

states_codes = [x.lower() for x in states['code'].tolist()]
states_names = [x.lower() for x in states['state'].tolist()]
cities_name = [str(x).lower() for x in cities['city'].tolist()]

counties_name = [str(x).lower() for x in counties['county'].tolist()]
```

```python
def get_states(location):
    
    location = str(location).lower()
    
    location = location.replace(' usa','')

    location = location.replace(' us','')

    location = location.replace(' the usa','')
    location = location.replace('the us','')    
    location = location.replace('United States', '')
    
#     print(location)
    

    location = location.replace('.', ' ')
    location = location.replace('/', ' ')
    location = location.replace(',', ' ')
    location = location.replace('|', ' ')
    
    if location[-1:] == ' ' or location[-1:]=='.':
        location = location[:-1]
    
    found_location = ''
    
    if ' ' in location:
#         more words
        #check state code
        if location in states_codes:
            found_location = location
            return location

        #check state name  # {'alabama': {'code': 'al'},
        if location in states_names:
            found_location = location
            location = state_to_code_mapping[location]['code']
            return location

        #check city name
        if location in cities_name:
            found_location = location
            location = cities_to_state_code_mapping[location]['code']
            return location  
        
            #check county name
        if location in counties_name:
                location = counties_to_state_code_mapping[location]['code']
                found_location = location
                return location        
        
        location_cells = location.split(' ')
        
        
        for location_cl in reversed(location_cells):
            if location_cl in states_codes:
                found_location = location_cl
                return location_cl

            #check state name
            if location_cl in states_names:
                found_location = location_cl
                location_cl = state_to_code_mapping[location_cl]['code']

                return location_cl

            #check city name
            if location_cl in cities_name:
                
                found_location = location_cl
                location_cl = cities_to_state_code_mapping[location_cl]['code']

                return location_cl             
            #check county name
            if location_cl in counties_name:
                location_cl = counties_to_state_code_mapping[location_cl]['code']
                found_location = location_cl
                return location_cl        
        
        #check combination of first two words combined 
        if len(location_cells) == 2:
            location_cl = ' '.join(location_cells[:2])
            
            if location_cl in states_codes:
                found_location = location_cl
                return location_cl

            #check state name
            if location_cl in states_names:
                found_location = location_cl
                location_cl = state_to_code_mapping[location_cl]['code']

                return location_cl

            #check city name
            if location_cl in cities_name:
                
                found_location = location_cl
                location = cities_to_state_code_mapping[location_cl]['code']

                return location_cl             
            #check county name
            if location_cl in counties_name:
                location_cl = counties_to_state_code_mapping[location_cl]['code']
                found_location = location_cl
                return location_cl             
        

    else:
#         one word possibly
        
        #check state code
        if location in states_codes:
            
            found_location = location
            return location

        #check state name
        if location in states_names:
            location = state_to_code_mapping[location]['code']
            
            found_location = location
            return location

        #check city name
        if location in cities_name:
            location = cities_to_state_code_mapping[location]['code']
            found_location = location
            return location
        
        #check city name
        if location in counties_name:
            location = counties_to_state_code_mapping[location]['code']
            found_location = location
            return location        

    return 'UNK'
```

```python
loc = []
for l in location:
    loc.append(get_states(l))
```
