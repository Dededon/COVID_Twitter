```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
retweet1 = pd.read_csv("retweet_10039.csv", error_bad_lines=False)
retweet2 = pd.read_csv("retweet_10040.csv", error_bad_lines=False)
```

```python
retweet = pd.concat([retweet1, retweet2], ignore_index=True)
retweet.head()
```

```python
file1 = pd.read_csv('combined_10039.csv')
file2 = pd.read_csv('combined_10040.csv')
file = pd.concat([file1, file2], ignore_index=True)
file.shape
```

```python
file.head()
```

```python
source_user = pd.concat([retweet['id_str'],retweet['user_id']], axis=1)
source_user.set_index('id_str')
```

```python
is_retweet = file['is_retweet']
user = file['user']
ori_id = file['original_tweet_id']
G = nx.Graph()
```

```python
user_dict = dict(source_user.values.tolist())
```

```python
for i in range(len(is_retweet)):
    if is_retweet[i]:
        G.add_node(user[i])
```

```python
for i in range(len(is_retweet)):
    if is_retweet[i]:
        if user_dict.get(ori_id[i])!=None:
            usid = user_dict.get(ori_id[i])
            G.add_node(usid)
            if G.has_edge(user[i], usid):
                G[user[i]][usid]['weight'] += 1
            else:
                G.add_edge(user[i],user_dict.get(ori_id[i]))
                G[user[i]][usid]['weight'] = 1
```

```python
G.remove_nodes_from(list(nx.isolates(G)))
G.number_of_nodes(), G.number_of_edges()
```
