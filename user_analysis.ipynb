```python
import pandas as pd
import os
```

```python
user_path = './data/users/fixed/'
user_files = os.listdir(user_path)
user_file = pd.read_csv(user_path + user_files[0], error_bad_lines=False)
for i in range(1,len(user_files)):
    temp = pd.read_csv(user_path + user_files[i], error_bad_lines=False)
    user_file = pd.concat([user_file, temp], ignore_index=True)
```

```python
cluster_result = pd.read_csv('.\output\mask_co_hashtag_t2_cluster.csv')
```

```python
user_file.head()
```

```python
for i in cluster_result['node']:
    print(i)
```

```python
user_dict = user_file.set_index('screen_name').T.to_dict('list')
```

```python
location = []
description = []
followers = []
name = []
created_at = []
likes = []
statuses_count = []
for screen_name in cluster_result['node']:
    data = user_dict.get(screen_name)
    location.append(data[1])
    description.append(data[2])
    followers.append(data[3])
    created_at.append(data[5])
    likes.append(data[6])
    name.append(data[7])
    statuses_count.append(data[13])
```

```python
cluster_result['location'] = location
cluster_result['description'] = description
cluster_result['followers'] = followers
cluster_result['name'] = name
cluster_result['created_at'] = created_at
cluster_result['likes'] = likes
cluster_result['statuses_count'] = statuses_count
```

```python
cluster_result.to_csv('.\output\cluster_result_features.csv')
```
